```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("csv_files/engineered/nfl_games_attendance.csv")
```

### Violin plot de capacidad vs asistencia

```python
plt.style.use("ggplot")

# plot
fig, ax = plt.subplots(figsize=(5, 5))
VP = ax.violinplot(
    pd.concat(
        [
            df["stadium_max_capacity"],
            df["stadium_regular_capacity"],
            df["game_attendance"],
        ],
        axis=1,
    ),
    widths=2,
    showmeans=False,
    showmedians=True,
    showextrema=False,
)

ax.set(ylim=(0, 100000))
ax.set_ylabel("People")
ax.legend()

# set style for the axes
labels = [
    "Average stadium \n max capacity",
    "Average stadium \n regular capacity",
    "Average game \n attendance",
]
ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
ax.set_xlim(0.25, len(labels) + 0.75)

plt.show()
```

### Pie plot de condiciones climaticas el dia de cada partido

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.pie(
    df["weather_condition"].value_counts(),
    labels=df["weather_condition"].unique(),
    radius=1,
    wedgeprops={"linewidth": 1, "edgecolor": "white"},
    autopct="%1.1f%%",
)

ax.set_title("Weather conditions frequency")

plt.show()
```
