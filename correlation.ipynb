```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("csv_files/engineered/nfl_games_attendance.csv")
```

### Generar df con los coeficientes de correlacion de cada variable con la variable asistencia

```python
# List of non-numeric columns that need to be converted into dummy variables
non_numeric_cols = [
    "winner",
    "day",
    "date",
    "time",
    "home_team_name",
    "away_team_name",
    "stadium_name",
    "home_city",
    "weather_condition",
]

# Initialize a new DataFrame to store the coefficients
df_coefficients = df.copy()

df_dummies = pd.get_dummies(df, prefix_sep="?", columns=non_numeric_cols, dtype="int")

attendance_corr = df_dummies.corr()["game_attendance"]
attendance_corr_keys = attendance_corr.keys()

for index, row_corr in enumerate(attendance_corr):
    row_name = attendance_corr_keys[index].split("?")[:-1]
    row_value = attendance_corr_keys[index].split("?")[-1]
    # print(row_name, row_value, row_corr)

    if row_name and row_name[0] in non_numeric_cols:
        df_coefficients.loc[
            df_coefficients[row_name[0]] == row_value, row_name[0] + "_corr"
        ] = row_corr

df_coefficients = df_coefficients[
    [
        "year",
        "week",
        "winner",
        "winner_corr",
        "day",
        "day_corr",
        "date",
        "date_corr",
        "time",
        "time_corr",
        "pts_win",
        "pts_loss",
        "home_team_name",
        "home_team_name_corr",
        "away_team_name",
        "away_team_name_corr",
        "game_attendance",
        "home_team_previous_year_performance",
        "away_team_previous_year_performance",
        "home_team_current_year_performance",
        "away_team_current_year_performance",
        "home_team_current_sos",
        "away_team_current_sos",
        "stadium_name",
        "stadium_name_corr",
        "stadium_max_capacity",
        "stadium_regular_capacity",
        "home_city",
        "home_city_corr",
        "weather_condition",
        "weather_condition_corr",
    ]
]
df_coefficients_non_numeric_cols = df_coefficients[
    [
        "winner",
        "winner_corr",
        "day",
        "day_corr",
        "date",
        "date_corr",
        "time",
        "time_corr",
        "home_team_name",
        "home_team_name_corr",
        "away_team_name",
        "away_team_name_corr",
        "stadium_name",
        "stadium_name_corr",
        "home_city",
        "home_city_corr",
        "weather_condition",
        "weather_condition_corr",
    ]
]


df_coefficients.to_csv("nfl_games_attendance_with_coefficients.csv", index=False)
```
